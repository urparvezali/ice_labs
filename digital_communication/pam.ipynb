### Pulse Amplitude Modulation
#### Theory
Pulse Amplitude Modulation (PAM) is a technique used in signal processing where the amplitude of a series of carrier pulses is varied in accordance with the amplitude of the message signal. In this lab, we generate a unipolar Non-Return-to-Zero (NRZ) signal as the carrier pulse train and modulate it with a sinusoidal message signal to produce the PAM signal.

The key steps involved are:
1. **Carrier Signal Generation**: A unipolar NRZ signal is created using the `unipolar_nrz` function. This signal acts as the carrier pulse train.
2. **Message Signal Generation**: A sinusoidal signal is generated to represent the analog message signal.
3. **Modulation**: The carrier signal is multiplied with the message signal to produce the PAM signal.

#### Objectives
- Understand the concept of Pulse Amplitude Modulation (PAM) and its applications.
- Generate a unipolar NRZ signal as a carrier pulse train using the `unipolar_nrz` function.
- Create a sinusoidal message signal to simulate an analog signal.
- Perform modulation by combining the carrier and message signals to produce the PAM signal.
- Visualize the carrier signal, message signal, and the resulting PAM signal using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def unipolar_nrz(data: np.ndarray, sampling_rate: int = 100, duration: float = 1):
    time = np.arange(0, len(data) * duration, duration / sampling_rate)
    signal = np.zeros_like(time)
    for i, bit in enumerate(data):
        start = i * sampling_rate
        signal[start : start + sampling_rate // 2] = bit
    return time, signal
```

```python
cf, mf, sr = 10, 1, 100
data = np.ones(40)
time, csignal = unipolar_nrz(data, sr, 1/cf)
msignal = np.sin(2*np.pi*mf*time)

modulated_signal = csignal * msignal
plt.figure(figsize=(8, 6))
plt.subplot(3, 1, 1)
plt.plot(time, msignal)
plt.title("Analog Signal")

plt.subplot(3, 1, 2)
plt.plot(time, csignal)
plt.title("Carrier Pulse Train")
plt.ylabel("Amplitude")

plt.subplot(3, 1, 3)
plt.plot(time, modulated_signal)
plt.title("PAM Signal")
plt.xlabel("Time (s)")

plt.tight_layout()
```
